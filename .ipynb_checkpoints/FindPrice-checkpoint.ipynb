```python
obj_col = train_df.select_dtypes(include=[object]).columns
#возвращает список со значением для каждой строковой колоны
def get_attrs(df, columns):
    all_attrs = {}
    for column in columns:
        all_attrs[column] = {}
        uval = df[column].unique()
        uval = [x for x in uval if not pd.isnull(x)]
            
        ind = 0
        for u in uval:
            all_attrs[column][u] = ind
            ind+=1
            
    return all_attrs

#меняет значения для атрибутов строковых колон(можно использовать get_dummies())
def change_type(df, columns):
    for column in columns:
        uval = df[column].unique()
        uval = [x for x in uval if not pd.isnull(x)]
        df[column] = df[column].replace(uval, [x for x in range(len(uval))])
        

my_attrs = get_attrs(train_df, obj_col)
change_type(train_df, obj_col)
```

```python
#Проверить коррелируемость атрибутов
for col in obj_col:
    train_df[col] = train_df[col].fillna(train_df[col].median())
```

# обработка

```python
train_df.insert(80, 'FlrNum', pd.Series([1 for x in range(train_df.shape[0])]))
```

```python
secflrndx = train_df.loc[train_df['2ndFlrSF']==0].index.copy()
train_df.loc[secflrndx, 'FlrNum'] = 2
train_df = train_df.drop(columns={'1stFlrSF', '2ndFlrSF'})
```

```python
train_df['AvgGarageArea'] = train_df['GarageArea']/train_df['GarageCars']
train_df['AvgRmsArea'] = train_df['GrLivArea']/train_df['TotRmsAbvGrd']
```

```python
afterfitch_drop = ['BsmtHalfBath', 'GarageCond', 'PoolQC', 'Electrical', 'Fence',
                    'RoofMatl', 'Alley', 'MasVnrType', 'Condition1', 'BsmtExposure',
                    'SaleType', 'PavedDrive', 'BsmtCond', 'EnclosedPorch', 'YrSold',
                    'LotConfig', 'BldgType', 'BsmtFullBath', 'GarageQual', 'HeatingQC',
                    'GarageFinish', 'Exterior1st', 'BsmtFinType2', 'FireplaceQu',
                    'RoofStyle', 'MoSold', 'LandContour', 'BsmtFinType1', 'Exterior2nd',
                    'FlrNum', 'Functional']
train_df = train_df.drop(columns={'GarageArea', 'GarageCars', 'TotRmsAbvGrd', 'GrLivArea', 'Id',
                                 'PoolArea', 'KitchenAbvGr', 'LowQualFinSF', 'BsmtFinSF2', '3SsnPorch',
                                 'ScreenPorch', 'MiscVal', 'BedroomAbvGr', 'Street', 'Utilities', 'Condition2',
                                 'Heating', 'MiscFeature', *afterfitch_drop})
```

```python
train_df.LotFrontage = train_df.LotFrontage.fillna(0)
train_df.AvgGarageArea = train_df.AvgGarageArea.fillna(0)
train_df.MasVnrArea = train_df.MasVnrArea.fillna(0)
train_df.GarageYrBlt = train_df.GarageYrBlt.fillna(train_df.GarageYrBlt.median())
```

# Модели

```python
#train_df = train_df.astype('float64')
y = train_df.SalePrice
X = train_df.drop(columns=['SalePrice'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
params = {'n_estimators':[x for x in range(200, 2000, 200)],
          'criterion':['friedman_mse'],
         'max_depth':[50, 100],
         'min_samples_split':[30]}

clf = RandomForestRegressor()
cv_clf = GridSearchCV(clf, params, cv=5)

cv_clf.fit(X_train, y_train)
```

```python
y_pred = cv_clf.best_estimator_.predict(X_test)
```

```python
print('rmse: %s'%round(mean_squared_error(y_test, y_pred), 3))
print('r2: %s'%round(r2_score(y_test, y_pred), 3))
```

```python
ftch_imp = pd.DataFrame({'feature':list(X_train),
                       'importance':cv_clf.best_estimator_.feature_importances_})\
                        .sort_values(by='importance', ascending=True)
```

```python
ftch_imp.plot(y='importance', x='feature', kind='barh', figsize=(12, 20))
```

```python
import joblib
```

```python
joblib.dump(cv_clf.best_estimator_, 'regressionforrest.joblib')
```
